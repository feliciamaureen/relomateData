```python
def removeNonNumerical(text):
    text = re.sub("[^0-9]", "", text)
    return text
```

```python
kosplus.isnull().sum()
```

```python
for i in range(len(prodUsers)):
    prodUsers.loc[i, 'Phone'] = removeNonNumerical(prodUsers.loc[i, 'Phone'])
```
